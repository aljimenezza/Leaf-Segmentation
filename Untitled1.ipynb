```python
# -*- coding: utf-8 -*-
"""
Created on Thu Jun 30 22:50:54 2022

@author: laura
"""

"""importar librerias"""
import numpy as np
#import plotly.graph_object as go
import matplotlib.pyplot as plt


"""
2.1 (10%). Simule la dinámica de L y Ca.  Grafique en una figura con 5 
páneles la dinámica de L, Ca, u3, u1 y q2.
"""

"""parametros"""
A = 7 						                 # Área del tanque [m^2]
Cv = 0.0005 			                     # Coeficiente de flujo [m^2/%]
q2 = 0.09					                 # Flujo de entrada 2 del tanque [m^3/s]
g=9.81                                       #Gravedad [m/s^2]
Ca1=25                                       #concentración del compuesto A en la corriente de alimentación 1 [mol/m^3]
Ca2=1                                        #concentración del compuesto A en la corriente de alimentación 2 [mol/m^3]


"""Condiciones iniciales"""
u3_o = 80					                 # Apertura inicial de la válvula 3 [%]
u1_o = 100					                 # Apertura inicial de la válvula 1 [%]
L_o = 0.5 					                 # Nivel inicial del tanque [m]
Ca_o = 0.9 					                 # Concentracion inicial de A [mol/m^3]

                                            
"""vector de tiempo"""
t_0=0                                        #tiempo inicial
t_f=800                                      #tiempo final
dt=0.01                                      #paso
iteraciones=round(t_f/dt)                      
t=np.linspace(t_0,t_f,(iteraciones+1))       #vector tiempo
                                             

"""método de euler"""
L_dif=np.zeros(iteraciones+1)
L=L_o                                        #nivel del tanque (m)
u_1=np.zeros(iteraciones+1)
u1=u1_o                                      #porcentaje de apertura de la válvula en la entrada 1 del tanque (%)
u_3=np.zeros(iteraciones+1)
u3=u3_o                                      #porcentaje de apertura de la válvula en la salida 3 del tanque (%)
q_2=np.zeros(iteraciones+1)
Ca_dif=np.zeros(iteraciones+1)               
Ca=Ca_o
                                             
for i in range(0, iteraciones+1):
    u_1[i]=u1                                
    u_3[i]=u3
    q_2[i]=q2
    q3=u3*Cv*np.sqrt(g*L)			         # Flujo de salida del tanque [m^3/s]
    q1=u1*0.09/100				             # Flujo de entrada 1 del tanque [m^3/s].  u1 = 100
    L_dif[i]=L
    dLdt=(q1/A)+(q2/A)-(q3/A)                #variación del nivel del tanque con el tiempo (dL/dt)
    L=L+(dLdt*dt)
    
    V=L*A                                  #volumen del líquido dentro del tanque (m^3)
    Ca_dif[i]=Ca
    dCadt=(1/V)*((q1*Ca1)+(q2*Ca2)-(q3*Ca))  #variación del componente A con el tiempo (dCa/dt)
    Ca=Ca+(dCadt*dt)


"""graficas"""
plt.figure(1)  #abre la figura 1

plt.subplot(5,1,1)  #grafica cinco filas y una columna en la primera posición
plt.plot(t,L_dif,'g',label='Dinámica de L')
plt.title('Dinámica de L')
plt.ylabel('Nivel del tanque [m]')

plt.legend(loc='best')
plt.grid("on")
plt.show()

plt.subplot(5,1,2)  #grafica cinco filas y una columna en la primera posición
plt.plot(t,Ca_dif,'b',label='Dinámica de Ca')
plt.title('Dinámica de Ca')
plt.ylabel('CA [mol/m^3]')

plt.legend(loc='best')
plt.grid("on")
plt.show()

plt.subplot(5,1,3)  #grafica cinco filas y una columna en la primera posición
plt.plot(t,u_3,'b',label='Gráfica de u3')
plt.title('u3')
plt.ylabel('% apertura V3')

plt.legend(loc='best')
plt.grid("on")
plt.show()

plt.subplot(5,1,4)  #grafica cinco filas y una columna en la primera posición
plt.plot(t,u_1,'b',label='Gráfica de u1')
plt.title('u1')
plt.ylabel('% apertura V1')

plt.legend(loc='best')
plt.grid("on")
plt.show()

plt.subplot(5,1,5)  #grafica cinco filas y una columna en la primera posición
plt.plot(t,q_2,'b',label='Gráfica de q2')
plt.title('Gráfica de q2')
plt.ylabel('flujo q2 [m^3/s]')
plt.xlabel('Tiempo [s]')
plt.legend(loc='best')
plt.grid("on")
plt.show()
```
